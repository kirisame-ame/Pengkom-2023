```python
import pandas as pd
import matplotlib as mat
df = pd.read_csv('P05-03.csv')
```

```python
#nomor 1
private = df.loc[df["room_type"]== "Private room"]
private[['number_of_reviews']].plot(kind="hist",bins=[i for i in range(0,131,30)],xticks= [i for i in range(0,131,30)])
```

```python
#nomor 2
good_rev = df.loc[df["number_of_reviews"] > 100]
good_rev["room_type"].value_counts().plot(kind="pie")
```

```python
#nomor 3
not_jay = df.loc[df["host_name"] != "Jay"]
not_jay.plot(kind="scatter",x="longitude",y="latitude")
```

```python
#nomor 4
novena = df.loc[df["neighbourhood"]== "Novena"]
novena.plot(kind="scatter",x="longitude",y="latitude",sizes=novena['availability_365'])
```

```python
#nomor 5
group = df.loc[(df["host_name"]== "Jojo") | (df["host_name"]== "Elizabeth") | (df["host_name"]== "Jimmy")]
group.groupby(["room_type","host_name"])["name"].size().unstack().plot(kind="bar",stacked=True)
```
